```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_excel("/content/Online retail.xlsx")
df.head()
```

```python
df.tail() # checkin for the last 5 variables
```

```python
df = df.rename(columns={'shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil': 'Transactions'})
df.head()
```

```python
print(df.isnull().sum())
```

```python
df = df.iloc[:, 0].dropna().apply(lambda x: x.split(',')).tolist()
```

```python
trans = TransactionEncoder()
trans_df = trans.fit_transform(df)
data=pd.DataFrame(trans_df, columns=trans.columns_,dtype='int')
data.head()
```

```python
data.isnull().sum() # checking for the null values
```

```python
data.info() # information of the data after transformation
```

```python
data.describe() # describing the data
```

## **Applying Aprior algorithm**

```python
frequent_itemsets = apriori(data, min_support=0.01, use_colnames=True)
frequent_itemsets
```

#**Applying Association rule**

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2,num_itemsets=3)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.confidence>=0.4]
```

## 1. Support (Frequency of an Itemset)

**Definition:** Support measures how frequently an item or itemset appears in all transactions.

**Why is it needed?**
- Helps filter out infrequent items that are not useful.
- Ensures we only consider items that appear frequently enough to be relevant.

---

## 2. Confidence (Reliability of the Rule)

**Definition:** Confidence measures how often an item **B** is bought when item **A** is also bought.

**Why is it needed?**
- Helps identify strong association rules.
- Higher confidence means the rule is more trustworthy.


---

## 3. Lift (Strength of Association)

**Definition:** Lift measures how much more likely item **B** is bought when **A** is bought, compared to buying **B** independently.

**Why is it needed?**
- **Lift > 1** → A and B are positively correlated (buying A increases chances of buying B).
- **Lift = 1** → A and B are independent (no relationship).
- **Lift < 1** → A and B are negatively correlated (buying A reduces chances of buying B).
