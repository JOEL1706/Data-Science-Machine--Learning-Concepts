```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
from scipy import stats
```

```python
data={'Smart Thermostat':[50,80,60,30,20],'Smart Light':[70,100,90,50,50]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
crosstable=pd.crosstab(df['Smart Thermostat'],df['Smart Light'])
```

```python
crosstable
```

```python
chi2_contingency(crosstable)
```

```python
chi2_stats,p_value,dof,expected=chi2_contingency(crosstable)
```

####critical value

```python
critical=stats.chi2.ppf(0.95,12)
critical
```

####Decision and conclusion

```python
if chi2_stats>critical:
    print('Reject Null Hypothesis,Smart Thermostat and Smart Light are dependent')
else:
    print('Accept Null Hypothesis,Smart Thermostat and Smart Light are independent')
```
